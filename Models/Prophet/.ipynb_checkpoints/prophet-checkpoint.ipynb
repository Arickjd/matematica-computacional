```python
# Realizando o split
train, val = series_weekly.split_before(0.83)
```

### Setup de treinamento utilizando OPTUNA 

```python
from prophet import Prophet  # Facebook's Prophet
```

```python
def objective(trial):
    global model

    # Parâmetros otimizáveis
    seasonal_periods = trial.suggest_int("seasonal_periods", 1, len(train))
    fourier_order = trial.suggest_int("fourier_order", 1, 15)

    # Inicializa o modelo Prophet
    model = Prophet()

    # Adiciona sazonalidade customizada
    model.add_seasonality(
        name='custom',
        period=seasonal_periods,
        fourier_order=fourier_order
    )

    # Treina o modelo
    model.fit(train)

    # Faz previsão
    future = model.make_future_dataframe(periods=len(val))
    forecast = model.predict(future)

    # Extrai apenas os valores previstos para o período de validação
    predicted = forecast[['ds', 'yhat']].tail(len(val))
    
    # Calcula o MAPE
    return mape(val['y'].values, predicted['yhat'].values)
```

```python
def callback(study,trial):
    global best_model
    if study.best_trial == trial:
        best_model = model
```

```python
study = optuna.create_study(direction = "minimize", storage=None)
study.optimize(objective, n_trials = 100, callbacks = [callback])
```
